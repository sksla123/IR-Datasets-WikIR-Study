# IR Dataset WikIR/en1k 데이터 셋 구조 분석

```python
dataset_name = "wikir/en1k"
iter_limits = 5
```

## WikIR/en1k 인덱스 분석

```python
import ir_datasets
from ir_datasets import registry
```

```python
wikir_superset = [key for key in registry._registered.keys() if key.startswith(dataset_name)]

print(wikir_superset)
```

## index별 구성 분석

### wikir/en1k

```python
wikir_en1k_dataset = ir_datasets.load("wikir/en1k")
dataset = wikir_en1k_dataset
```

```python
# 구성요소
dataset.metadata()
```

```python
# docs 분석
for counter, doc in enumerate(dataset.docs_iter()):
    if counter == 0:
        print(type(doc))
    print(doc)
    if counter > iter_limits - 1: 
        break
```
